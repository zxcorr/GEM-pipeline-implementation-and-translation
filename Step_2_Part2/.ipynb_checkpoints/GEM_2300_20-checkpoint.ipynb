```python
def recr_gain_rd_bsl_corr(azimut,azoff):
    """
    
    """
    
    #dentro do loop de linhas de um arquivo
    #receiver gain and radiometer baseline corrections
    
    azimut += azoff
    if azimut < 0: azimut += 360
    if azimut > 360: azimut += -360
    
    # celsius => Kelvin
    DT2 = T2 - T2cal + 273.15 
      
    T1 = T1 + 273.15 # radiometers sensors
    T2 = T2 + 273.15
    T4 = T4 + 273.15
         
    Gaint = Gain/(etagain*DT2 + 1)
    dG1 = Gaint/Gain  
    dG2 = Gaint*etagain/(etagain*DT2 + 1) 
    
    backgr2	= TAgnd30*T4/300. + TAatm30 + TAeg + TAcbr 
    
    if mode == 1:
        TA = signal*Gaint - Tsys
    else:
        TA = signal
        
    if jgnd == 1:
        TA = TA - azibin(int(azimut/3)+1)
    
    dTA = math.sqrt((signal*dG1 * dGain)**2 + (signal*dG2*dT2cal)**2) 
    
    #decidir return
    return     
```

```python
def sun_ephemeris_2(T, LS, RS, BS, jday, month, jyear, hour):
    """
    
    """
    #heliocentric coordinates of the sun
    L, B, R, IMODE = 0, 0, 0, 0
    
    X, Y, Z, DELTA0 = geocen(T, L, B, R, LS, BS, RS, IMODE)
    
    X, Y, Z = eclequ(T, X, Y, Z)
    
    X, Y, Z = nutequ(T, X, Y, Z)
    
    delta, dec, RA = polar(X, Y, Z)
    
    RA = RA/15
    
    tau = lmst(mjd(jday,month,jyear,hour)) - RA
    
    if tau < 0.00000001:
        tau += 24
        
    tau *= 15
    
    LAT, H, AZ = equhor(dec, tau)
    
    return LAT, H, AZ
```

```python
def mjd(id, mh, iy, hr):
    """
    Calcula a Data Juliana Modificada (MJD) a partir do dia, mês, ano e hora.
    
    Parâmetros:
    - id: Dia
    - mh: Mês
    - iy: Ano
    - hr: Hora (em formato decimal)
    
    Retorna:
    - MJD: Data Juliana Modificada
    """
    # Ajusta o mês e ano se o mês for janeiro ou fevereiro
    if mh <= 2:
        mh += 12
        iy -= 1
    
    # Ajusta para calendário juliano/gregoriano
    A = 10000.0 * iy + 100.0 * mh + id
    
    if A <= 15821004.1:  # Antes de 15 de outubro de 1582 (calendário juliano)
        jB = -2 + (iy + 4716) // 4 - 1179
    else:  # Após 15 de outubro de 1582 (calendário gregoriano)
        jB = iy // 400 - iy // 100 + iy // 4
    
    # Cálculo final da MJD
    A = 365.0 * iy - 679004.0
    mjd = A + jB + floor(30.6001 * (mh + 1)) + id + hr / 24.0
    
    return mjd    
```

```python
def FRAC(x):
    """
    Calcula a parte fracionária de um número.
    
    Parâmetros:
    - x: Número real (float)
    
    Retorna:
    - Parte fracionária de x no intervalo [0, 1)
    """
    x = x - floor(x)  # Subtrai a parte inteira (equivale ao AINT no Fortran)
    
    if x < 0.0:  # Ajusta para intervalo [0, 1)
        x = x + 1.0
    
    return x
```

```python
def addthe(c1,s1,c2,s2,c,s):
    """
    
    """
    C = c1*c2 - s1*s2
    S = s1*c2 + c1*s2
    
    return C,S
```

```python
def term(I1, I, IT, DLC, DLS, DRC, DRS, DBC, DBS):
    """
    
    """
    global U, V, C3, S3, C, S, DL, DR, DB
    
    DL = DR = DB = 0.0

    if IT == 0:
        addthe(C3[index(I1)], S3[index(I1)], C[index(I)], S[index(I)], U, V)
    else:
        U *= T
        V *= T

    # Atualização das variáveis DL, DR e DB
    DL += DLC * U + DLS * V
    DR += DRC * U + DRS * V
    DB += DBC * U + DBS * V

    return DL, DR, DB, C, S, U, V
```

```python
def geocen(T, LP, BP, RP, LS, BS, RS, IMODE):
    """
    Calcula coordenadas geocêntricas.
    
    Args:
        T (float): Tempo em séculos julianos desde J2000.
        LP, BP, RP (float): Coordenadas eclípticas heliocêntricas do planeta.
        LS, BS, RS (float): Coordenadas eclípticas geocêntricas do Sol.
        IMODE (int): Tipo de coordenada desejada (0=geométrica, 1=astrométrica, 2=aparente).

    Returns:
        tuple: Coordenadas finais (X, Y, Z, DELTA0)
    """
    P2 = 8 * math.atan(1.0)
    
    DL = DB = DR = 0.0
    DLS = DBS = DRS = 0.0
    
    if IMODE > 0:
        M = P2 * frac(0.9931266 + 99.9973604 * T)
        DLS = 172.00 + 5.75 * math.sin(M)
        DRS = 2.87 * math.cos(M)
        DBS = 0.0
    
    XS, YS, ZS, VXS, VYS, VZS = posvel(LS, BS, RS, DLS, DBS, DRS)
    XP, YP, ZP, VX, VY, VZ = posvel(LP, BP, RP, DL, DB, DR)

    X = XP + XS
    Y = YP + YS
    Z = ZP + ZS
    
    DELTA0 = math.sqrt(X**2 + Y**2 + Z**2)
    
    FAC = 0.00578 * DELTA0 * 1e-4

    X -= FAC * (VX + VXS)
    Y -= FAC * (VY + VYS)
    Z -= FAC * (VZ + VZS)
    
    return X, Y, Z, DELTA0
```

```python
def posvel(L, B, R, DL, DB, DR):
    """
    Converte coordenadas esféricas em cartesianas e calcula velocidades.
    
    Args:
        L (float): Longitude em graus.
        B (float): Latitude em graus.
        R (float): Raio (distância).
        DL (float): Variação na longitude.
        DB (float): Variação na latitude.
        DR (float): Variação no raio.
        
    Returns:
        tuple: (X, Y, Z, VX, VY, VZ) coordenadas cartesianas e velocidades.
    """
    Pi = math.pi
    CL = math.cos(L * Pi / 180)
    SL = math.sin(L * Pi / 180)
    CB = math.cos(B * Pi / 180)
    SB = math.sin(B * Pi / 180)
    
    X = R * CL * CB
    VX = DR * CL * CB - DL * R * SL * CB - DB * R * CL * SB
    Y = R * SL * CB
    VY = DR * SL * CB + DL * R * CL * CB - DB * R * SL * SB
    Z = R * SB
    VZ = DR * SB + DB * R * CB
    
    return X, Y, Z, VX, VY, VZ
```

```python
def eclequ(T,X,Y,Z):
    """
    
    """
    # Conversion of ecliptic into equatorial coordinates       
    # (equinox in Julian centuries since J2000)     
    pi = math.pi
    
    EPS = 23.43929111-((46.8150+(0.00059-0.001813*T)*T)*T)/3600.0
    C = math.cos(EPS*Pi/180.)  
    S = math.sin(EPS*Pi/180.)

    V = C*Y - S*Z  
    Z = S*Y + C*Z  
    Y = V
    
    return V, Y, Z
```

```python
def nutequ(t, x, y, z):
    """
    Applies corrections to Cartesian coordinates (x, y, z) due to nutation.
    Parameters:
        t (float): Time variable.
        x, y, z (float): Cartesian coordinates.
    Returns:
        tuple: Corrected Cartesian coordinates (x, y, z).
    """
    arc = 206264.8062  # arcseconds per radian
    p2 = 8 * math.atan(1.0)  # pi

    ls = p2 * FRAC(0.993133 + 99.997306 * t)  # mean anomaly Sun
    d = p2 * FRAC(0.827362 + 1236.853087 * t)  # diff. longitude Moon-Sun
    f = p2 * FRAC(0.259089 + 1342.227826 * t)  # mean argument of latitude
    n = p2 * FRAC(0.347346 - 5.372447 * t)  # longitude of ascending node
    EPS = 0.4090928 - 2.2696e-4 * t  # obliquity of the ecliptic

    dpsi = (
        -17.200 * math.sin(n)\
        - 1.319 * math.sin(2 * (f - d + n))\
        - 0.227 * math.sin(2 * (f + n))\
        + 0.206 * math.sin(2 * n)\
        + 0.143 * math.sin(ls)) / arc

    deps = (
        9.203 * math.cos(n)\
        + 0.574 * math.cos(2 * (f - d + n))\
        + 0.098 * math.cos(2 * (f + n))\
        - 0.090 * math.cos(2 * n)\) / arc

    C = dpsi * math.cos(EPS)
    S = dpsi * math.sin(EPS)

    dx = -(c * y + s * z)
    dy = c * x - deps * z
    dz = s * x + deps * y

    x += dx
    y += dy
    z += dz

    return x, y, z
```

```python
def polar(x, y, z):
    """
    Converts Cartesian coordinates (x, y, z) to polar coordinates (r, theta, phi).
    Parameters:
        x, y, z (float): Cartesian coordinates.
    Returns:
        tuple: Polar coordinates (r, theta, phi).
    """
    rho = x**2 + y**2
    r = math.sqrt(rho + z**2)
    phi = math.atan2(y, x)

    if phi < 0:
        phi += 2 * math.pi  # Convert to range [0, 2*pi]

    rho = math.sqrt(rho)
    theta = math.atan2(z, rho)

    # Convert phi and theta from radians to degrees
    phi = math.degrees(phi)
    theta = math.degrees(theta)

    return r, theta, phi
```

```python
def lmst(fmjd, lambd):
    """
    Calculate the Local Mean Sidereal Time (LMST).

    Parameters:
        fmjd (float): Modified Julian Date (fractional).
        lambd (float): Longitude in degrees (positive East).

    Returns:
        float: Local Mean Sidereal Time (in hours).
    """

    mjd0 = math.floor(fmjd)
    ut = (fmjd - mjd0) * 24.0
    t = (mjd0 - 51544.5) / 36525.0

    gmst = (
        6.697374558
        + 1.0027379093 * ut
        + (8640184.812866 + (0.093104 - 6.2e-6 * t) * t) * t / 3600.0
    )

    aux = (gmst - lambd / 15.0) / 24.0
    lmst = 24.0 * FRAC(aux)

    return lmst
```

```python
def equhor(dec, tau, phi):
    """
    Converts equatorial coordinates into horizontal coordinates.

    Parameters:
        dec (float): Declination in degrees (-90 to +90).
        tau (float): Hour angle in degrees (0 to 360).
        phi (float): Geographical latitude in degrees.

    Returns:
        tuple: Altitude (H) in degrees, Azimuth (AZ) in degrees (0 to 360, counted S->W->N->E->S).
    """
    pi = math.pi

    # Precompute trigonometric functions
    cs_phi = math.cos(math.radians(phi))
    sn_phi = math.sin(math.radians(phi))
    cs_dec = math.cos(math.radians(dec))
    sn_dec = math.sin(math.radians(dec))
    cs_tau = math.cos(math.radians(tau))

    # Cartesian coordinates
    x = cs_dec * sn_phi * cs_tau - sn_dec * cs_phi
    y = cs_dec * math.sin(math.radians(tau))
    z = cs_dec * cs_phi * cs_tau + sn_dec * sn_phi

    r, h, az = polar(x, y, z)
    
    return r, h, az
```
